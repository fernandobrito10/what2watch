```python
import pandas as pd

df = pd.read_csv('/content/watchlist.csv')

df
```

```python
import requests
from bs4 import BeautifulSoup
import re

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"}

for index, row in df.iterrows():
  url = row['Letterboxd URI']
  response = requests.get(url, headers=headers)
  soup = BeautifulSoup(response.text, 'html.parser')
  try:
    rating = soup.find("meta", {"name": "twitter:data2"})
    duration = soup.find("p", class_="text-link text-footer")
    mins = re.search(r'(\d+)\xa0mins', duration.get_text())
    df.at[index, 'Rating'] = rating["content"][:4]
    df.at[index, 'Duration'] = mins.group(1)
  except Exception as e:
    print("Error:", e)
df
```
